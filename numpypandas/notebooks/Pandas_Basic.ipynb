# Pandas Series

```python
# Pandas Series is a one-dimensional labeled array capable of holding data of any type 
# (integer, string, float, python objects, etc.).

# series a list or from numpy array 
print('Series from list')
print(pd.Series([1, 2, 3, 4]))
print('\nSeries from numpy 1d array')
print(pd.Series(np.array([1, 2, 3, 4])))
print('by default an index beginning from 0 is created')
print('\n creating a pandas series with a defined index')
series_with_def_index = pd.Series([1, 2, 3, 4], index=['one,', 'two', 'three', 'four'])
print(series_with_def_index )
print('The series elements can be accessed with array indices or the defined index')
print(series_with_def_index[1])
print(series_with_def_index['two'])
print(series_with_def_index.two)
print('\nSeries from a dictioinary - the keys will be the index')
series_fm_dict = pd.Series({0:0, 1:1, 2:4, 3:9, 4:16})
print(series_fm_dict)
print(series_fm_dict[0])
```

# Filtering Dataframes

```python
# filtering on conditions
mask = sample_df.matches > 30
sample_df[mask]
```

```python
frequent_and_scoring = (sample_df.matches > 20) & (sample_df.centuries > 5)
sample_df[frequent_and_scoring]
```

```python
matches_or_innings = (sample_df.matches > 20) | (sample_df.innings > 40)
sample_df[matches_or_innings]
```

```python
mask_for_np = np.logical_or(sample_df.matches > 20, sample_df.innings > 40)
sample_df[mask_for_np]
```

```python
global_mask = sample_df > 10
sample_df[global_mask]
```

# Data mainpulation
## modifying values, adding, deleting rows, columns

```python
sample_df_copy = sample_df.copy()
sample_df_copy.loc['rohan', 'matches'] = 21
print(sample_df_copy.loc['rohan', 'matches'])
print(sample_df.loc['rohan', 'matches'])
```

```python
# sample_df_copy[sample_df_copy.matches == 20]["innings"] = 35
# A value is trying to be set on a copy of a slice from a DataFrame.
# Try using .loc[row_indexer,col_indexer] = value instead

# See the caveats in the documentation: 
# https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy
sample_df_copy[sample_df_copy.matches == 20]["innings"] = 35
sample_df_copy[['matches', 'innings']]
```
