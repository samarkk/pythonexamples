```python
import numpy as np
import pandas as pd
```

```python
from helper import load_sample_data_from_dir
```

```python
fut_df_data_location = 'D:/findataf/fo_sample_data'
fut_df = load_sample_data_from_dir(fut_df_data_location)
# drop the extra column due to trailing comma in the source files
fut_df.drop(['Unnamed: 15'],axis=1,inplace=True)
fut_df
```

```python
futdfgr = fut_df.groupby(['INSTRUMENT', 'SYMBOL', 'OPTION_TYP'], as_index=False).agg(
contracts=('CONTRACTS', 'sum'))
futdfgr
```

```python
futdfgr.pivot(index=['INSTRUMENT', 'SYMBOL'],columns='OPTION_TYP',
              values='contracts')
```

```python
# use the pandas pivot table functionality which offers more options
pd.pivot_table(fut_df, values=['CONTRACTS'], columns=['OPTION_TYP'], 
               aggfunc=['sum', 'mean'], index=['SYMBOL', 'EXPIRY_DT'] )
```

```python
# to flatten hierarchical columns lets assign to a variable
# margins = true wull lead to total aggregations across both axes
# fut_df pivoted
ftdpvt =pd.pivot_table(fut_df, values=['CONTRACTS'], columns=['OPTION_TYP'], 
                       aggfunc='sum', index=['SYMBOL', 'EXPIRY_DT'], margins=True )

# flatten heirarchical column names
new_col_names = [x[0]+'_'+x[1] for x in ftdpvt.columns]
ftdpvt.columns = new_col_names
ftdpvt
```

```python
# #### melt
'''
Unpivot a DataFrame from wide to long format, optionally leaving identifiers set.

This function is useful to massage a DataFrame into a format where one
or more columns are identifier variables (`id_vars`), while all other
columns, considered measured variables (`value_vars`), are "unpivoted" to
the row axis, leaving just two non-identifier columns, 'variable' and
'value'.
'''
smalldf = fut_df.loc[(fut_df.SYMBOL == 'ITC') & (fut_df.INSTRUMENT == 'FUTSTK')]
# to see more rows
pd.options.display.max_rows = 500
smalldf.melt(
    id_vars=['INSTRUMENT', 'SYMBOL'],
    value_vars=['CONTRACTS', 'CHG_IN_OI'],)
```

```python
"""
## Crosstab
Like pivot_table but for arrays not dataframes. And uses frequency not mean as default
"""
subsetdf = fut_df.loc[(fut_df.SYMBOL == 'ITC') | (fut_df.SYMBOL == 'NTPC') |
                      (fut_df.SYMBOL == 'AXISBANK')]
pd.crosstab(index=subsetdf['INSTRUMENT'],columns=subsetdf['SYMBOL'])
```

```python
pd.crosstab(index=[subsetdf['INSTRUMENT'], subsetdf['SYMBOL']],
            columns=subsetdf['EXPIRY_DT'],aggfunc='sum',values=['CONTRACTS'])
```
