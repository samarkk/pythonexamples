```python
cgrpured = credit.groupby(['purpose', 'employment_duration'])
cgrpured.sum(numeric_only=True)
```

```python
sqldf('''
select purpose, employment_duration,  
avg(months_loan_duration) as avg_mld,
avg(amount) as avg_amount from credit
group by  purpose, employment_duration
''')
```

```python
# if we want to group by a continuous variable we can use cuts to transform it into a categorical one
amount_bins = [int(250 + (x * 500)) for x in range(38)]
amount_bins = [250, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 15000, 20000]
cuts = pd.cut(credit.amount, amount_bins, include_lowest = True)
credit['amountgr'] = cuts
credit.groupby(['purpose', 'amountgr']).sum(numeric_only=True)
```

```python
# aggregations
# use dictionary
credit.groupby('housing').agg({
    'amount': ['mean', 'max', 'min', 'std']
})
```

```python
credit.groupby(['purpose', 'savings_balance']).agg({
    'amount': ['mean', 'max', 'min', 'std'],
    'age': 'mean'
})
```

```python
# use dictionary and tuples to assign names to aggregation columns
credit.groupby(['purpose', 'savings_balance']).agg({
    'amount': [('count', 'count'),('amtmean', 'mean'), ('amtmax', 'max'), 
                ('amtmin', 'min'), ('amtstd', 'std')],
    'age': 'mean'
})
```

```python
# use reset_index() to  include grouping columns in data frame
credit.groupby(['purpose', 'savings_balance']).agg({
    'amount': [('count', 'count'),('amtmean', 'mean'), ('amtmax', 'max'), 
                ('amtmin', 'min'), ('amtstd', 'std')],
    'age': 'mean'
}).reset_index()
```

```python
# if we want to apply customized function
def variance(x):
    return np.std(x) / np.sqrt(x.size)
```

```python
# use tuples and assignment to use specific function and name the aggregated column
credit.groupby(['purpose', 'savings_balance']).agg(
    avg_amount=('amount', 'mean'),
    var_amount=('amount', variance)
)
```

# Handling missing values, outliers ####

```python
# read the credit file again
credit_file_loc = 'D:/localmts/pythonml/credit_msng_mdfd.csv'
credit = pd.read_csv(credit_file_loc)
# display the rows with null values
credit[credit.isnull().any(axis=1)]
```

```python
crdrop = credit.dropna()
```

```python
crdrop[crdrop.isnull().any(axis=1)]
```

```python
print(credit.shape)
print(crdrop.shape)
```

```python
# fill the amount na values with  the mean amount values
credit['amount'] = credit['amount'].transform( lambda x : x.fillna(x.mean()))
print(credit.loc[966].amount)
print(credit.loc[1000].amount)
```

```python
print(credit.amount.mean())
print(credit.groupby(['purpose','credit_history'])['amount'].mean())
```

```python
# fill the amount na values as per attributes
credit['amount'] = credit.groupby(['purpose','credit_history'])['amount'].transform(
     lambda x : x.fillna(x.mean()))
print(credit.loc[966].amount)
print(credit.loc[1000].amount)
```

```python
credit['years_at_residence'] = credit.groupby(['purpose','credit_history'])['years_at_residence'].transform(
    lambda x : x.fillna(x.mean()))
```

```python
print(credit.years_at_residence.mean())
print(credit.groupby(['purpose','credit_history'])['years_at_residence'].mean().loc[('car', 'critical')])
```

```python
credit['savings_balance'] = credit.groupby('checking_balance')['savings_balance'].transform(
    lambda x : x.fillna(x.value_counts().index[1]))
```

```python
# gr4sd, lt4sd - greater than, less than 4 standard deviations
gr4sd = (credit.months_loan_duration > credit.months_loan_duration.mean() + 4 *  credit.months_loan_duration.std())
lt4sd = (credit.months_loan_duration <  credit.months_loan_duration.mean() - 4 *  credit.months_loan_duration.std())
```

```python
credit[gr4sd | lt4sd]
```

```python
credit_outlier_indices = credit[gr4sd | lt4sd].index
```
