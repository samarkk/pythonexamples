```python
'''
 Apply a function along an axis of the DataFrame.

    Objects passed to the function are Series objects whose index is
    either the DataFrame's index (``axis=0``) or the DataFrame's columns
    (``axis=1``). By default (``result_type=None``), the final return type
    is inferred from the return type of the applied function. Otherwise,
    it depends on the `result_type` argument.
'''
print(fruits_quality_df.jan.apply(np.log))
print(fruits_quality_df.jan.apply(lambda x: x + 100))
```

```python
def doubleIfLT100(x):
    if (x < 100):
        return x * 2
    return x
```

```python
%%timeit
fruits_quality_df.jan.apply(doubleIfLT100)
# fruits_quality_df.loc[:,['jan']]
# fruits_quality_df.loc[:,['jan', 'feb', 'mar', 'apr', 'may', 'jun']].apply(doubleIfLT100, axis=1)
```

```python
# vectorized function execution will be faster
def vecDouble(x):
    x[x < 100] *= 2
    return x
```

```python
%%timeit
fruits_quality_df.loc[:,['jan']].apply(vecDouble)
```
